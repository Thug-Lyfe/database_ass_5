```python
%sql select * from chinook.artist,chinook.album where ((chinook.artist.artistid = chinook.album.artistid) OR (chinook.artist.name like '%Bach%' OR chinook.artist.name like '%Mozart%')) AND (chinook.album.title like '%Bach%' OR chinook.album.title like '%Mozart%')
```

```python
%sql select * from chinook.track where composer like '%Bach%' OR composer like '%Mozart%'
```

```python
%sql select album.title from chinook.track,chinook.album where (album.albumid = track.albumid) AND (track.composer like '%Mozart%' OR track.composer like '%Bach%')
```
